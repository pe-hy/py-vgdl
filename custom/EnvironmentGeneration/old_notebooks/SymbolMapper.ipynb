```python
import pickle
import os
```

```python
pickles = os.path.abspath('../pickles')
```

```python
with open(os.path.join(pickles, 'definition_dic.pkl'), 'rb') as file:
    definition_dic = pickle.load(file)
```

```python
OBSTACLE_INCR = 1000
```

```python
definition_dic.keys()
```

```python
obstacle_dic = {k: v["final_sym"]+OBSTACLE_INCR for k, v in definition_dic["mol_keys_dic"].items()}
```

```python
with open(os.path.join(pickles, 'obstacle_dic.pkl'), 'wb') as file:
    pickle.dump(obstacle_dic, file)
```

```python
all_def_dic = definition_dic["all_dic"]
all_def_dic
```

```python
mol_def_dic = definition_dic["mol_keys_dic"]
```

```python
final_sym_lst = []
for k, v in mol_def_dic.items():
    final_sym = v["final_sym"]
    final_sym_lst.append(final_sym)
```

```python
number_set = set()
for key, value in all_def_dic.items():
    number_set.add(key[0])
    number_set.add(key[1])
    number_set.add(value)

print(number_set)
```

```python
symbols = [chr(i) for i in range(945, 970)] + [chr(i) for i in range(1000, 1200)]
symbols_for_obstacles = [i+OBSTACLE_INCR for i in final_sym_lst]
symbols_sorted = sorted(symbols)

number_set.update(symbols_for_obstacles)
symbol_mappings = {}

for i, number in enumerate(sorted(number_set)):
    symbol_mappings[number] = symbols_sorted[i]
print(symbol_mappings)
```

```python
with open(os.path.join(pickles, 'symbol_mappings.pkl'), 'wb') as file:
    pickle.dump(symbol_mappings, file)
```
